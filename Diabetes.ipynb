```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train_sc,y_train)
```

```python
y_pred = dt.predict(X_test_sc)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score
```

```python
print(classification_report(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
from sklearn.metrics import roc_auc_score
```

```python
oversampler = SMOTE()
```

```python
X_train_sc, y_train = oversampler.fit_resample(X_train_sc, y_train)
```

```python
X_test_sc, y_test = oversampler.fit_resample(X_test_sc, y_test)
```

```python
print(X_train_sc.shape)
print(y_train.shape)
```

```python
dt.fit(X_train_sc, y_train)
```

```python
y_pred = dt.predict(X_test_sc)
```

```python
print(classification_report(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
print(roc_auc_score(y_pred, y_test))
```

# Cross Validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_folds = 5
```

```python
param_grid = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : range(5,50,5),
    'min_samples_leaf' : range(5,50,5),
    'min_samples_split' : range(5,50,5)
}
```

```python
cv = GridSearchCV(estimator=dt,
                 cv=n_folds,
                 verbose=3,
                 n_jobs=-1,
                param_grid = param_grid,
                 return_train_score=True)
```

```python
cv.fit(X_train_sc, y_train)
```

```python
cv.best_score_
```

```python
cv.best_estimator_
```

```python
dt_final = DecisionTreeClassifier(max_depth=20, min_samples_leaf=5, min_samples_split=20)
```

```python
dt_final.fit(X_train_sc, y_train)
```

```python
y_pred = dt_final.predict(X_test_sc)
```

```python
print(classification_report(y_pred, y_test))
```

```python
print(roc_auc_score(y_pred, y_test))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train_sc, y_train)
```

```python
y_pred_tr = rfc.predict(X_train_sc)
y_pred_ts = rfc.predict(X_test_sc)
```

```python
print(accuracy_score(y_pred_tr, y_train))
print(accuracy_score(y_pred_ts, y_test))
```

```python
n_folds = 3
param_grid = {
    'criterion': ['gini','entropy'],
    'max_depth': range(5,50,10),
    'min_samples_leaf': range(5,50,10),
    'min_samples_split': range(5,50,10),
    'n_estimators': [100,200],
    'max_features': ['sqrt','log2']
}
```

```python
cv = GridSearchCV(
    estimator=rfc,
    n_jobs=-1,
    verbose=3,
    cv=n_folds,
    param_grid=param_grid
)
```

```python
cv.fit(X_train_sc, y_train)
```

```python
cv.best_estimator_
```

```python
rfc_final = RandomForestClassifier(criterion='entropy', max_depth=45, max_features='sqrt',
                       min_samples_leaf=5, min_samples_split=15,
                       n_estimators=200)
```

```python
rfc_final.fit(X_train_sc, y_train)
```

```python
y_pred_rfc_final = rfc_final.predict(X_test_sc)
```

```python
print(classification_report(y_pred_rfc_final, y_test))
```

```python
print(confusion_matrix(y_pred_rfc_final, y_test))
```

```python
print(roc_auc_score(y_pred_rfc_final, y_test))
```
